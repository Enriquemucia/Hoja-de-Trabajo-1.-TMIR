```python
def activate_color_channels(img_path, color):
   
    b, g, r = cv2.split(img)
    
    zeros = np.zeros(img.shape[:2], dtype="uint8")
    
    if color == 1:
        resolucion = cv2.merge([b, zeros, zeros])
    elif color == 2:
        resolucion = cv2.merge([zeros, g, zeros])
    elif color == 3:
        resolucion = cv2.merge([zeros, zeros, r])
    elif color == 10:
        resolucion = cv2.merge([zeros, g, r])
    elif color == 20:
        resolucion = cv2.merge([b, g, zeros])
    elif color == 30:
        resolucion = cv2.merge([b, zeros, r])
    else:
        raise ValueError("Color debe ser 1, 2, 3, 10, 20, o 30")
    
    return resolucion
```

```python
def create_3d_grayscale(image_path, output_prefix):
    img = cv2.imread(image_path)
    
    if img is None:
        raise ValueError("La imagen no se pudo cargar.")
    
    b, g, r = cv2.split(img)
    
    cv2.imwrite(f'{output_prefix}_gray_azul.jpg', b)
    cv2.imwrite(f'{output_prefix}_gray_verde.jpg', g)
    cv2.imwrite(f'{output_prefix}_gray_rojo.jpg', r)

image_path = 'img3.JPG'
output_prefix = 'img3_salida'

create_3d_grayscale(image_path, output_prefix)
```

### Problema No. 4

```python
def plot_histograms(image_path):
    img = cv2.imread(image_path)
    
    if img is None:
        raise ValueError("La imagen no se pudo cargar.")
    
    b, g, r = cv2.split(img)
    
    gray = np.mean(img, axis=2).astype(np.uint8)
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    colors = ('blue', 'green', 'red')
    channels = (b, g, r)
    titles = ('Histogram for Blue Channel', 'Histogram for Green Channel', 'Histogram for Red Channel', 'Histogram for Grayscale')
    
    for ax, color, channel, title in zip(axes.flatten()[:3], colors, channels, titles[:3]):
        hist = cv2.calcHist([channel], [0], None, [256], [0, 256])
        ax.plot(hist, color=color)
        mean_val = np.mean(channel)
        ax.axvline(mean_val, color=color, linestyle='dashed', linewidth=1)
        ax.set_title(title)
        ax.set_xlim([0, 256])
        ax.legend([f'Mean: {mean_val:.2f}'])
    
    ax = axes.flatten()[3]
    hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
    ax.plot(hist, color='gray')
    mean_val = np.mean(gray)
    ax.axvline(mean_val, color='gray', linestyle='dashed', linewidth=1)
    ax.set_title(titles[3])
    ax.set_xlim([0, 256])
    ax.legend([f'Mean: {mean_val:.2f}'])
    
    plt.tight_layout()
    plt.show()

image_path = 'img4.jpg'
plot_histograms(image_path)
```

### Enfoque de escala de grises ponderado

El enfoque de escala de grises ponderado consiste en asignar diferentes pesos a los canales de color (rojo, verde y azul) para calcular una imagen en escala de grises. Esta tcnica se utiliza porque el ojo humano no percibe todos los colores con la misma intensidad; generalmente, el verde es percibido con mayor intensidad que el rojo, y el rojo ms que el azul.

Una frmula comnmente utilizada para la conversin ponderada a escala de grises es:
Grayscale=0.299 x Red + 0.587 x Green + 0.114 x Blue


```python
def weighted_grayscale(image_path):
    img = cv2.imread(image_path)
    
    if img is None:
        raise ValueError("La imagen no se pudo cargar.")
    
    b, g, r = cv2.split(img)
    
    gray = 0.299 * r + 0.587 * g + 0.114 * b
    gray = gray.astype(np.uint8)
    
    return gray

image_path = 'img5.jpg'
gray_image = weighted_grayscale(image_path)

plt.imshow(gray_image, cmap='gray')
plt.axis('off')  # Ocultar los ejes
plt.show()

cv2.imwrite('img5_gray.jpg', gray_image)
```

### Espacio de Color HSV
El espacio de color HSV es un modelo de representacin de color que describe los colores en trminos de sus componentes de tono, saturacin y valor. Este modelo es frecuentemente utilizado en aplicaciones de grficos por computadora y procesamiento de imgenes porque se asemeja ms a la manera en que los humanos perciben y describen los colores.

#### Componentes del Espacio HSV
#### Hue (H) - Tono:

Representa el tipo de color (como rojo, verde, azul, etc.).
Se mide en grados en un crculo cromtico de 0 a 360.
Por ejemplo, 0 es rojo, 120 es verde, y 240 es azul.

#### Saturation (S) - Saturacin:
Representa la pureza del color (la cantidad de gris en el color).
Se mide en porcentaje de 0% a 100%.
Un color con 0% de saturacin es un tono de gris, mientras que un color con 100% de saturacin es un color puro sin gris.

#### Value (V) - Valor (Brillo):
Representa el brillo del color.
Se mide en porcentaje de 0% a 100%.
Un color con 0% de valor es completamente negro, mientras que un color con 100% de valor es tan brillante como el color puede ser.
